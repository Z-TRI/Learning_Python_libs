```python
car_sales["Make"]= car_sales["Make"].str.lower()
car_sales
```

```python
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
car_sales_missing["Odometer"].mean(numeric_only=True)
```

```python
# Method 1: Fill missing values with mean using fillna() inplace=True
car_sales_missing.fillna({"Odemeter": car_sales_missing["Odometer"].mean()}
                                     , inplace=True)
car_sales_missing
```

```python
# Method 2: Fill missing values with mean using fillna() by assign dataframe itself
car_sales_missing["Odometer"]= car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(numeric_only=True))
car_sales_missing
```

```python
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
```

```python
# Ignore all missing values in the dataframe
car_sales_missing_dropped= car_sales_missing.dropna()
car_sales_missing_dropped
```

```python
car_sales= pd.read_csv("car-sales.csv")
```

```python
# Columns from series
seats_column = pd.Series([5, 5, 5, 5, 5])

# New column from seats
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales.fillna({"Seats": 5}, inplace=True)
car_sales
```

```python
# Columns from Python list
# list must be same length as columns in dataframe
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5]
car_sales["Fuel per 100KM"]= fuel_economy
car_sales
```

```python
car_sales["Total fuel used (L)"]= car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
car_sales
```

```python
# create a column from a signle value
car_sales["Number of wheels"]= 4
car_sales
```

```python
car_sales["Passed road safety"]= True
car_sales
```

```python
car_sales = car_sales.drop("Total fuel used", axis=1)
car_sales
```

```python
# randomly shuffle the data
car_sales_shuffled = car_sales.sample(frac=1)
car_sales_shuffled
```

```python
# Only select 20% of the data
car_sales_shuffled.sample(frac=0.2)
```

```python
car_sales_shuffled.reset_index(drop=True, inplace=True)
car_sales_shuffled
```
